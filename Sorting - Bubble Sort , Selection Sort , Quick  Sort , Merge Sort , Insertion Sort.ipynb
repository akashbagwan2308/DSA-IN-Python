# Bubble Sort

```python
def BubbleSort(arr):
    for i in range(len(arr)):
        for j in range(len(arr)-i-1):
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
    return arr
```

```python
arr = [7,8,1,3,2]
BubbleSort(arr)
```

# Selection Sort

```python
def SelectionSort(arr):
    for i in range(len(arr)):
        curr = i
        for j in range(i,len(arr)):
            if arr[curr]>arr[j]:
                curr = j
        arr[curr],arr[i]=arr[i],arr[curr]
    return arr
```

```python
arr = [7,8,1,3,2]
SelectionSort(arr)
```

# Quick Sort

```python
def QuickSort(arr):
    if len(arr)<=1:
        return arr
    pivot = arr.pop()
    greater = []
    lower = []
    for i in arr:
        if i > pivot:
            greater.append(i)
        else:
            lower.append(i)
    return QuickSort(lower) + [pivot] + QuickSort(greater)
```

```python
arr = [7,8,1,3,2]
QuickSort(arr)
```

# Merge Sort

```python
def merge(arr,a1,a2):
    i=j=k=0
    while i<len(a1) and j< len(a2):
        if a1[i]<a2[j]:
            arr[k]=a1[i]
            i+=1
            k+=1
        else:
            arr[k]=a2[j]
            k+=1
            j+=1
    while i < len(a1):
        arr[k]=a1[i]
        k+=1
        i+=1
    while j< len(a2):
        arr[k]=a2[j]
        k+=1
        j+=1
    return arr
```

```python
def MergeSort(arr):
    if len(arr)<=1:
        return arr
    mid = len(arr)//2
    a1 = arr[:mid]
    a2 = arr[mid:]
    MergeSort(a1)
    MergeSort(a2)
    return merge(arr,a1,a2)
```

```python
arr = [7,8,1,3,2]
MergeSort(arr)
```

# Insertion Sort

```python
def InsertionSort(arr):
    for i in range(1,len(arr)):
        current = arr[i]
        j = i-1
        while j>=0 and current <arr[j]:
            arr[j+1]=arr[j]
            j-=1
        arr[j+1]=current
    return arr
```

```python
arr = [7,8,1,3,2]
InsertionSort(arr)
```
