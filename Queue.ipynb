## Queue using array

```python
class Queue:
    def __init__(self):
        self.__arr = []
        self.__front = 0
        self.__count = 0
        self.__idx = 0
        self.__rear = -1
    def enqueue(self,data): 
        self.__arr.append(data)
        self.__count += 1
        self.__rear += 1
    def dequeue(self):
        if self.__count == 0:
            return -1
        ele = self.__arr[self.__front]
        self.__front += 1
        self.__count -=1
        return ele
    def front(self):
        if self.__count == 0:
            return -1
        return self.__arr[self.__front]
    def size(self):
        return self.__count
    def isEmpty(self):
        return self.size() == 0
    def printQueue(self):
        self.__idx = self.__front
        while self.__idx <=self.__rear:
            print(self.__arr[self.__idx], end = ' ')
            self.__idx += 1
```

```python
q = Queue()
q.enqueue(5)
q.enqueue(2)
q.enqueue(4)
q.enqueue(15)
q.enqueue(51)
```

```python
q.printQueue()
```

```python
q.dequeue()
q.printQueue()
```

```python
q.dequeue()
q.printQueue()
```

## Queue using Linked LIst : 

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
```

```python
class QueueLL(Node):
    def __init__(self):
        self.__head = None
        self.__tail = None
        self.__count = 0
    def enQueue(self,element):
        newNode = Node(element)
        if self.__head is None:
            self.__head = newNode
            self.__tail = newNode
            self.__count +=1
        else:
            self.__tail.next = newNode
            self.__tail = self.__tail.next
            self.__count +=1
    def deQueue(self):
        if self.__head is not None:
            idx = self.__head
            self.__head = self.__head.next
            self.__count -=1
            return idx.data
        else:
            print("Queue is Empty!!")
            return
    def size(self):
        return self.__count
    
    def isEmpty(self):
        return self.size()==0
    def front(self):
        if self.isEmpty():
            print("Queue is Empty!!")
            return
        return self.__head.data
    def printQueue(self):
        id = self.__head 
        while id is not None:
            print(id.data,end =" ")
            id = id.next
```

```python
q = QueueLL()
q.enQueue(5)
q.enQueue(1)
q.enQueue(7)
q.enQueue(9)
q.enQueue(11)
```
