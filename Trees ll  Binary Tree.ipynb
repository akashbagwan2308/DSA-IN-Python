# Trees / Binary Tree

```python
class BTNode:
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None
```

```python
def printTree(root):
    if root == None:
        return
    print(root.data)
    printTree(root.left)
    printTree(root.right)
```

#### ~ Much Effective printTree Solution :

```python
def printTree1(root):
    if root == None:
        return
    print(root.data,end = ":")
    if root.left != None:
        print("L",root.left.data,end=",")
    if root.right != None:
        print("R",root.right.data,end = "")
    print()
    printTree1(root.left)
    printTree1(root.right)
```

```python
Bt1 = BTNode(1)
Bt2 = BTNode(4)
Bt3 = BTNode(5)
```

```python
Bt1.left = Bt2
Bt1.right = Bt3
```

### Input to the BinaryTree :

```python
def treeInput():
    rootData = int(input())
    if rootData == -1:
        return None
    root = BTNode(rootData)
    leftTree = treeInput()
    rightTree = treeInput()
    root.left = leftTree
    root.right = rightTree
    return root
```

```python
bt = treeInput()
```

```python
printTree1(bt)
```

### Number Of Nodes :

```python
def numNodes(root):
    if root is None:
        return 0
    leftCount = numNodes(root.left)
    rightCount = numNodes(root.right)
    return 1+leftCount+rightCount
```

```python
print(numNodes(bt))
```

### LargeNode:

```python
def largeNode(root):
    if root == None:
        return -1
    leftLarge = largeNode(root.left)
    rightLarge = largeNode(root.right)
    return max(root.data,leftLarge,rightLarge)
```

```python
print(largeNode(bt))
```

### Height of Tree :

```python
def treeHeight(root):
    if root is None:
        return 0
    if root.left is None and root.right is None:
        return 1
    leftHeight = treeHeight(root.left)
    rightHeight = treeHeight(root.right)
    return 1 + max(leftHeight,rightHeight)
```

```python
treeHeight(bt)
```

### Number of Leaf :

```python
def totalLeaf(root):
    if root is None:
        return 0
    if root.left is None and root.right is None:
        return 1
    leftleaf = totalLeaf(root.left)
    rightleaf = totalLeaf(root.right)
    return leftleaf + rightleaf
```

```python
bt = treeInput()
printTree1(bt)
totalLeaf(bt)
```

### PrintNodes at depth k :

```python
def printNode(root,k):
    if root is None:
        return
    if k == 0:
        print(root.data)
        return
    printNode(root.left,k-1)
    printNode(root.right,k-1)   
```

```python
bt = treeInput()
printTree1(bt)
printNode(bt,2)
```

```python
printNode(bt,1)
```

```python
def printNode1(root,k,d):
    if root is None:
        return
    if k==d:
        print(root.data)
        return
    printNode1(root.left,k,d+1)
    printNode1(root.right,k,d+1)
```

```python
printNode1(bt,2,0)
```

### Remove Leaf 

```python
def removeLeaf(root):
    if root is None:
        return
    if root.left is None and root.right is None:
        return
    root.left = removeLeaf(root.left)
    root.right = removeLeaf(root.right)
    return root
```

```python
bt = treeInput()
printTree1(bt)
```

```python
removeLeaf(bt)
printTree1(bt)
```

### IsTreeBalanced 

```python
def isTreeBalanced(root):
    if root is None:
        return 0
    if root.left is None and root.right is None:
        return 1
    return isTreeBalanced(root.left)==isTreeBalanced(root.right)
```

```python
bt = treeInput()
printTree1(bt)
isTreeBalanced(bt)
```

### IsTreeBalanced Improved

```python
def isBalanced(root):
    if root is None:
        return 0,True
    lh, isleftbalanced = isBalanced(root.left)
    rh,isrightbalanced = isBalanced(root.right)
    h = 1+ max(lh,rh)
    if lh-rh > 1 or rh-lh > 1 :
        return h , False
    if isleftbalanced and isrightbalanced:
        return h, True
    else:
        return h, False
```

```python
print(isBalanced(bt))
```

### Diameter of tree 

```python
def diameter(root):
    if root is None:
        return 0
    op1 = treeHeight(root.left)+ treeHeight(root.right)
    op2 = diameter(root.left)
    op3 = diameter(root.right)
    return max(op1,op2,op3)
```

```python
print(diameter(bt))
```

### Level Wise Input 

```python
import queue
def LWInput():
    q = queue.Queue()
    print("Enter root :")
    rootData = int(input())
    if rootData == -1:
        return None
    root = BTNode(rootData)
    q.put(root)
    while not q.empty():
        currNode = q.get()
        print("Enter left child of ",currNode.data ,":")
        leftchildData = int(input())
        if leftchildData != -1:
            leftNode = BTNode(leftchildData)
            currNode.left = leftNode
            q.put(leftNode)
            
        print("Enter right child of ",currNode.data ,":")
        rightchildData = int(input())
        if rightchildData != -1:
            rightNode = BTNode(rightchildData)
            currNode.right = rightNode
            q.put(rightNode)
    return root    
```

```python
Bt = LWInput()
printTree1(Bt)
```

### PrintTree LEvel Wise 

```python
def printLW(root):
    q = queue.Queue()
    q.put(root)
    while not q.empty():
        a = q.get()
        if a == None:
            break
        print(a.data,end=" ")
        if a.left is not None or a.right is not None:
            q.put(a.left)
            q.put(a.right)
```

```python
printLW(Bt)
```

### Build tree using Inorder & Preorder

```python
class BTNode:
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None   
```

```python
def buildTree(preOrder,inOrder):
    if inOrder:
        root = BTNode(preOrder.pop(0))
        root_index = inOrder.index(root.data)
        root.left = buildTree(preOrder,inOrder[:root_index])
        root.right = buildTree(preOrder,inOrder[root_index+1:])
        return root
```

```python
printTree1(buildTree([3,9,20,15,7],[9,3,15,20,7]))
```

### Build tree using Inorder & Postorder

```python
def buildtree(postOrder,inOrder):
    if inOrder:
        root = BTNode(postOrder.pop())
        root_index = inOrder.index(root.data)
        root.right = buildtree(postOrder,inOrder[root_index+1:])
        root.left = buildtree(postOrder,inOrder[:root_index])
        return root
```

```python
printTree1(buildtree([9,15,7,20,3],[9,3,15,20,7]))
```
