# Generic Tree

Generic trees are a collection of nodes where each node is a data structure that consists of records and a list of references to its children(duplicate references are not allowed). Unlike the linked list, each node stores the address of multiple nodes. Every node stores address of its children and the very first node’s address will be stored in a separate pointer called root.

### Generic Tree Node Class

```python
class GTNode:
    def __init__(self,data):
        self.data = data
        self.children = list()
```

### Print Generic Tree

```python
def printGT(root):
    if root is None:
        return
    print(root.data)
    for child in root.children:
        printGT(child)
```

### Print Generic Tree Improved

```python
def printGT1(root):
    if root is None:
        return
    print(root.data,":",end="")
    for child in root.children:
        print(child.data,",",end="")
    print()
    for child in root.children:
        printGT1(child)
```

### Input to GT

```python
def gtInput():
    print("Enter root data")
    rootData = int(input())
    if rootData == -1:
        return None
    root = GTNode(rootData)
    print("Enter no. of children for ",rootData)
    childrencount = int(input())
    for i in range(childrencount):
        child = gtInput()
        root.children.append(child)
    return root
```

### Level wise input 

```python
import queue
def lwInput():
    q = queue.Queue()
    print("Enter root data :")
    rootData = int(input())
    if rootData == -1:
        return None
    root = GTNode(rootData)
    q.put(root)
    while (not(q.empty())):
        current_node = q.get()
        print("Enter no. of children of ",current_node.data)
        numChild = int(input())
        for i in range(numChild):
            print("Enter next child for ",current_node.data)
            childData = int(input())
            child = GTNode(childData)
            current_node.children.append(child)
            q.put(child)
    
    return root          
    
```

### No. of Nodes

```python
def numNodes(root):
    if root is None:
        return 0
    count = 1
    for child in root.children:
        count = count + numNodes(child)
        
    return count
```
