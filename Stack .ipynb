# Stack  and Its Application :

### Stack USing Arrays/List :

```python
class Stack:
    def __init__(self):
        self.__ele = []
    def Push(self,data):
        self.__ele.append(data)
    def Pop(self):
        if self.isEmpty():
            print("Stack is empty!!!")
            return
        return self.__ele.pop()
    def Top(self):
        if self.isEmpty():
            print("Stack is empty!!!")
            return
        return self.__ele[-1]
    def isEmpty(self):
        return len(self.__ele)==0
    def Size(self):
        return len(self.__ele)
    def printStack(self):
        for elements in self.__ele:
            print(elements)
```

```python
s1 = Stack()
s1.Push(5)
s1.Push(25)
s1.Push(3)
s1.Push(9)
s1.Push(27)
```

```python
s1.printStack()
```

```python
s1.printStack()
```

### Stack using Linked List :

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
```

```python
class Stack1(Node):
    def __init__(self):
        self.__head = None
        self.__count = 0
    def Push(self,element):
        newNode = Node(element)
        newNode.next = self.__head
        self.__head = newNode
        self.__count +=1
    def Pop(self):
#         if self.isEmpty() is True:
#             print("Stack is Empty!!!")
#             return
        data = self.__head.data
        self.__head = self.__head.next
        self.__count -=1
        return data
    def Top(self):
        if self.isEmpty() is True:
            print("Stack is Empty!!!")
            return
        data = self.__head.data
        return data        
    def Size(self):
        return self.__count
    def isEmpty(self):
        return self.Size()==0
    def printStack(self):
        tail = self.__head
        while tail is not None:
            print(tail.data)
            tail = tail.next
```

### Valid Parenthesis

```python
def isbalanced(st):
    s = []
    for char in st:
        if char in '({[':
            s.append(char)
        elif char == ')':
            if (not s or s[-1]!= '('):
                return False
            s.pop()
        elif char == '}':
            if (not s or s[-1]!= '{'):
                return False
            s.pop()
        elif char == ']':
            if (not s or s[-1]!= '['):
                return False
            s.pop()
    if not s:
        return True
    return False
```

## Reverse Stack 

```python
def reverseStack(s1,s2):
    if len(s1)<=1:
        return
    while len(s1)!=1:
        ele = s1.pop()
        s2.append(ele)
    lastele = s1.pop()
    
    while(len(s2)!=0):
        ele = s2.pop()
        s1.append(ele)
    reverseStack(s1,s2)
    s1.append(lastele)
    return s1
```
