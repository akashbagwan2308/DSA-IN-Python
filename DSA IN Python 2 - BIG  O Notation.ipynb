# Big O Notation 

big O notation is used to measure how running time or space requirements for your program grow as input size grows

#####  Rules for Big O Notation : ------ 

1. Keep fastest growing term
2. Drop constant


------------------------------------------------------
1. Measuring Running time growth ---> Time complexity
2. Measuring Space Growth ----> Space complexity

```python
# Example 1
# for this proogram
# time = O(n)
def get_squared_number(number):
    squared_numbers = []
    for n in numbers:
        squared_numbers.append(n*n)
    return squared_numbers
```

```python
numbers = [2,5,8,9]
get_squared_number(numbers)
```
